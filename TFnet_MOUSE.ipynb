```python
def Badis_ungapped_TF_binding_sites(file_handle):
    return pd.read_csv(file_handle, delimiter='\t',\
                       names=['sequence', 'complement',\
                              'median_sig_int_1', 'e_score_1', 'z_score_1',\
                              'median_sig_int_2', 'e_score_2', 'z_score_2',\
                              'P_value_z_score_1', 'P_value_z_score_2', 'P_value_z_score_12',\
                              'P_value_e_score_1', 'P_value_e_score_2', 'P_value_e_score_12',\
                              'Q_value_z_score_1', 'Q_value_z_score_2', 'Q_value_z_score_12',\
                              'Q_value_e_score_1', 'Q_value_e_score_2', 'Q_value_e_score_12'],\
                       index_col='sequence')
```

```python
files_to_ignore = ['Bhlhb2_4971_contig8mers.txt', 'Gata3_4971_contig8mers.txt',\
                   'Rfx3_4971_contig8mers.txt', 'Sox7_4971_contig8mers.txt']

with zipfile.ZipFile('Badis_etal_ungapped.zip', 'r') as zf:
    tf_bind_site = dict()
    for name in zf.namelist():
        if name not in files_to_ignore:
            with zf.open(name, 'r') as f:
                tf_bind_site[name.split('_')[0]] = Badis_ungapped_TF_binding_sites(f)

for df in tf_bind_site.values():
    df['e_score'] = 0.5 * (df['e_score_1'] + df['e_score_2'])
```

## SNPs network

```python
import networkx as nx
```

```python
def snps(seed, alphabet, template=None):

    seqs = set()

    if template:
        i = 0
        while True:
            i = template.find('1', i)
            if i != -1:
                template = '0'.join([template[:i], template[i + 1:]])
                seqs |= {l.join([seed[:i], seed[i + 1:]]) for l in alphabet if l != seed[i]}
            else:
                break

    else:
        for i,s in enumerate(seed):
            seqs |= {l.join([seed[:i], seed[i + 1:]]) for l in alphabet if l != s}

    return seqs
```

```python
cutoff = 0.35

tf_graph = dict()
for tf in tf_bind_site.keys():
    tf_graph[tf] = nx.empty_graph()

    for seq in tf_bind_site[tf].index:
        if tf_bind_site[tf].ix[seq].e_score >= cutoff:

            tf_graph[tf].add_node(seq, e_score=tf_bind_site[tf].ix[seq].e_score)
            for mut in snps(seq, 'ACGT'):
                if mut in tf_graph[tf]:
                    tf_graph[tf].add_edge(seq, mut)
```

Count of the number of TFs that can bind to the non-neutral neighbours of a node.
*e.g.* There is only a single non-neutral neighbor. It binds to both TF1 (e-score = 0.45) and TF2 (e-score = 0.40). Since both of them can bind to this neighbor (e-score > 0.35), K = 2.

```python
for tf in tf_graph.keys():
    P = set()
    for node in tf_graph[tf]:
        # snps - neighbours
        nn_neighs = snps(node, 'ACGT') - set(tf_graph[tf][node])
        
        K = 0
        for other_tf in tf_graph.keys():
            if other_tf != tf:
                for seq in nn_neighs:
                    if seq in tf_graph[other_tf]:
                        K += 1
                        P.add(other_tf)
                        continue

        tf_graph[tf].node[node]['K'] = K
    tf_graph[tf].graph['P'] = len(P)
```

Count the number of TFs that can bind to the non-neutral neighbours of a node. If a neighbour can bind multiple TFs, only the most strongly bound is counted.
*e.g.* There is only a single non-neutral neighbor. It binds to both TF1 (e-score = 0.45) and TF2 (e-score = 0.40). When this neighbor would become available by mutations, it would bind only TF1, K = 1.

```python
size, nnets, q, K, P = list(), list(), list(), list(), list()

for g in tf_graph.values():
        size.append(g.number_of_nodes())
        nnets.append(len(nx.connected_component_subgraphs(g)))

        q.append(np.average(g.degree().values()) / 24.0)
        K.append(np.average([g.node[node]['K'] for node in g]) / 103.0)
        P.append(g.graph['P'] / 103.0)

set_stats = pd.DataFrame(np.vstack((size, nnets, q, K, P)).T, columns=['size', 'nnets', 'q', 'K', 'P'])
set_stats.head()
```
