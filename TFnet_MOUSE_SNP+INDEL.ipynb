## SNP+INDELs network

```python
import networkx as nx
```

```python
def snps(seed, alphabet, template=None):

    seqs = set()

    if template:
        i = 0
        while True:
            i = template.find('1', i)
            if i != -1:
                template = '0'.join([template[:i], template[i + 1:]])
                seqs |= {l.join([seed[:i], seed[i + 1:]]) for l in alphabet if l != seed[i]}
            else:
                break

    else:
        for i,s in enumerate(seed):
            seqs |= {l.join([seed[:i], seed[i + 1:]]) for l in alphabet if l != s}

    return seqs
```

```python
def inserts_helper(seed, max_gap_len=1):
    
    seqs = set()
    l = len(seed)

    for x in range(l + 1):
        for i in range(1, max_gap_len + 1):
            in_seed = seed[:x] + i * '.' + seed[x:]
            if in_seed[:l] != seed:
                seqs.add(in_seed[:l])
            if in_seed[-l:] != seed:
                seqs.add(in_seed[-l:])

    return seqs
```

```python
def deletions_helper(seed, max_gap_len=1):

    seqs = set()
    l = len(seed)

    for x in range(l):
        for i in range(1, max_gap_len + 1):
            if x + i <= l:
                del_seed = seed[:x] + seed[x + i:]
                seqs.add(i * '.' + del_seed)
                seqs.add(del_seed + i * '.')

    return seqs
```

```python
def indels(seed, alphabet, max_gap_len=1):

    seqs = set()
    
    for seq in inserts_helper(seed, max_gap_len=max_gap_len) | deletions_helper(seed, max_gap_len=max_gap_len):
        dotted = {seq}
        # enough replacing rounds to remove all '.'s from seq
        for replace_round in range(seq.count('.')):
            __dotted = set()
            for __seq in dotted:
                # replace a single '.' by all the letter in the alphabet
                __dotted |= {__seq.replace('.', letter, 1) for letter in alphabet}
            dotted = __dotted
        seqs |= dotted

    return seqs - {seed}
```

```python
cutoff = 0.35

tf_graph_snp, tf_graph_indel = dict(), dict()
for tf in tf_bind_site:
    tf_graph_snp[tf] = nx.empty_graph()
    tf_graph_indel[tf] = nx.empty_graph()

    for seq in tf_bind_site[tf].index:
        if tf_bind_site[tf].ix[seq].e_score >= cutoff:

            tf_graph_snp[tf].add_node(seq, e_score=tf_bind_site[tf].ix[seq].e_score)
            tf_graph_indel[tf].add_node(seq, e_score=tf_bind_site[tf].ix[seq].e_score)

            snp_muts = snps(seq, 'ACGT')
            tf_graph_snp[tf].node[seq]['k_snp'] = len(snp_muts)
            for mut in snp_muts:
                if mut in tf_graph_snp[tf]:
                    tf_graph_snp[tf].add_edge(seq, mut)

            indel_muts = indels(seq, 'ACGT')
            tf_graph_indel[tf].node[seq]['k_indel'] = len(indel_muts)
            for mut in indel_muts:
                if mut in tf_graph_indel[tf]:
                    tf_graph_indel[tf].add_edge(seq, mut)
```

```python
for tf in tf_bind_site:
    
    for node in tf_graph_snp[tf]:
        tf_graph_snp[tf].node[node]['q_snp'] = tf_graph_snp[tf].degree(node)
        
    P = set()
    for node in tf_graph_snp[tf]:
        # snps - neighbours
        nn_neighs = snps(node, 'ACGT') - set(tf_graph_snp[tf][node])
        
        K = 0
        for other_tf in tf_graph_snp:
            if other_tf != tf:
                for seq in nn_neighs:
                    if seq in tf_graph_snp[other_tf]:
                        K += 1
                        P.add(other_tf)
                        continue

        tf_graph_snp[tf].node[node]['K_snp'] = K
    tf_graph_snp[tf].graph['P_snp'] = len(P)
    
    for node in tf_graph_indel[tf]:
        tf_graph_indel[tf].node[node]['q_indel'] = tf_graph_indel[tf].degree(node)
        
    P = set()
    for node in tf_graph_indel[tf]:
        # indels - neighbours
        nn_neighs = indels(node, 'ACGT') - set(tf_graph_indel[tf][node])
        
        K = 0
        for other_tf in tf_graph_indel:
            if other_tf != tf:
                for seq in nn_neighs:
                    if seq in tf_graph_indel[other_tf]:
                        K += 1
                        P.add(other_tf)
                        continue

        tf_graph_indel[tf].node[node]['K_indel'] = K
    tf_graph_indel[tf].graph['P_indel'] = len(P)
```

```python
tf_graph = dict()
for tf in tf_bind_site:
    tf_graph[tf] = nx.MultiGraph(P_snp=tf_graph_snp[tf].graph['P_snp'], P_indel=tf_graph_indel[tf].graph['P_indel'])
    
    for node in tf_graph_snp[tf]:
        tf_graph[tf].add_node(node, e_score=tf_graph_snp[tf].node[node]['e_score'],\
                              k_snp=tf_graph_snp[tf].node[node]['k_snp'],\
                              q_snp=tf_graph_snp[tf].node[node]['q_snp'],\
                              K_snp=tf_graph_snp[tf].node[node]['K_snp'])
    for node in tf_graph_indel[tf]:
#         assert(node in tf_graph[tf])
        tf_graph[tf].node[node]['k_indel'] = tf_graph_indel[tf].node[node]['k_indel']
        tf_graph[tf].node[node]['q_indel'] = tf_graph_indel[tf].node[node]['q_indel']
        tf_graph[tf].node[node]['K_indel'] = tf_graph_indel[tf].node[node]['K_indel']

    for edge in tf_graph_snp[tf].edges():
        tf_graph[tf].add_edge(edge[0], edge[1], mutation='snp')
    for edge in tf_graph_indel[tf].edges():
        tf_graph[tf].add_edge(edge[0], edge[1], mutation='indel')
```

```python
bhlhs = ['Bhlhb2', 'Max', 'Myf6', 'Tcf3']
```

```python
cutoff = 0.35

bhlh_graph_snp, bhlh_graph_indel = dict(), dict()
for tf in tf_bind_site:
    if tf in bhlhs:
        bhlh_graph_snp[tf] = nx.empty_graph()
        bhlh_graph_indel[tf] = nx.empty_graph()

        for seq in tf_bind_site[tf].index:
            if tf_bind_site[tf].ix[seq].e_score >= cutoff:

                bhlh_graph_snp[tf].add_node(seq, e_score=tf_bind_site[tf].ix[seq].e_score)
                bhlh_graph_indel[tf].add_node(seq, e_score=tf_bind_site[tf].ix[seq].e_score)

                snp_muts = snps(seq, 'ACGT')
                bhlh_graph_snp[tf].node[seq]['k_snp'] = len(snp_muts)
                for mut in snp_muts:
                    if mut in bhlh_graph_snp[tf]:
                        bhlh_graph_snp[tf].add_edge(seq, mut)

                indel_muts = indels(seq, 'ACGT')
                bhlh_graph_indel[tf].node[seq]['k_indel'] = len(indel_muts)
                for mut in indel_muts:
                    if mut in bhlh_graph_indel[tf]:
                        bhlh_graph_indel[tf].add_edge(seq, mut)
```

```python
for tf in tf_bind_site:
    if tf in bhlhs:

        for node in tf_graph_snp[tf]:
            bhlh_graph_snp[tf].node[node]['q_snp'] = bhlh_graph_snp[tf].degree(node)

        P = set()
        for node in bhlh_graph_snp[tf]:
            # snps - neighbours
            nn_neighs = snps(node, 'ACGT') - set(bhlh_graph_snp[tf][node])

            K = 0
            for other_tf in bhlh_graph_snp:
                if other_tf != tf:
                    for seq in nn_neighs:
                        if seq in bhlh_graph_snp[other_tf]:
                            K += 1
                            P.add(other_tf)
                            continue

            bhlh_graph_snp[tf].node[node]['K_snp'] = K
        bhlh_graph_snp[tf].graph['P_snp'] = len(P)

        for node in bhlh_graph_indel[tf]:
            bhlh_graph_indel[tf].node[node]['q_indel'] = bhlh_graph_indel[tf].degree(node)

        P = set()
        for node in bhlh_graph_indel[tf]:
            # indels - neighbours
            nn_neighs = indels(node, 'ACGT') - set(bhlh_graph_indel[tf][node])

            K = 0
            for other_tf in bhlh_graph_indel:
                if other_tf != tf:
                    for seq in nn_neighs:
                        if seq in bhlh_graph_indel[other_tf]:
                            K += 1
                            P.add(other_tf)
                            continue

            bhlh_graph_indel[tf].node[node]['K_indel'] = K
        bhlh_graph_indel[tf].graph['P_indel'] = len(P)
```

```python
bhlh_graph = dict()
for tf in tf_bind_site:
    if tf in bhlhs:

        bhlh_graph[tf] = nx.MultiGraph(P_snp=bhlh_graph_snp[tf].graph['P_snp'], P_indel=bhlh_graph_indel[tf].graph['P_indel'])

        for node in bhlh_graph_snp[tf]:
            bhlh_graph[tf].add_node(node, e_score=bhlh_graph_snp[tf].node[node]['e_score'],\
                                  k_snp=bhlh_graph_snp[tf].node[node]['k_snp'],\
                                  q_snp=bhlh_graph_snp[tf].node[node]['q_snp'],\
                                  K_snp=bhlh_graph_snp[tf].node[node]['K_snp'])
        for node in bhlh_graph_indel[tf]:
    #         assert(node in bhlh_graph[tf])
            bhlh_graph[tf].node[node]['k_indel'] = bhlh_graph_indel[tf].node[node]['k_indel']
            bhlh_graph[tf].node[node]['q_indel'] = bhlh_graph_indel[tf].node[node]['q_indel']
            bhlh_graph[tf].node[node]['K_indel'] = bhlh_graph_indel[tf].node[node]['K_indel']

        for edge in bhlh_graph_snp[tf].edges():
            bhlh_graph[tf].add_edge(edge[0], edge[1], mutation='snp')
        for edge in bhlh_graph_indel[tf].edges():
            bhlh_graph[tf].add_edge(edge[0], edge[1], mutation='indel')
```
